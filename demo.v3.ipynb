# 1D example

## source data

```python
data = simulation.SimulatedData( N_sec=4 )
```

```python
# plot stimuli
fig = plt.figure()
ax = fig.gca( xlabel='time', ylabel='position', xticks=[], yticks=[], title='stimuli' )
ax.imshow( data.X__td.T, cmap=plt.cm.Greys )

# plot receptive field
fig = plt.figure()
ax = fig.gca( xlabel='position', ylabel='value', xticks=[], yticks=[], title='receptive field' )
ax.plot( data.k_stim_true, lw=2 )
pass
```

## solving

```python
# maximum likelihood solution
ml = glm.ML( data, testing_proportion=0.2 )
ml.solve( verbose=False )

# ridge penalised solution
ridge = glm.Ridge( data, testing_proportion=0.2 )
ridge.training_slices = ml.training_slices
ridge.testing_slices = ml.testing_slices
ridge.solve( verbose=False )
```

```python
# plot receptive fields
fig = plt.figure()
ax = fig.gca( xlabel='position', ylabel='value', xticks=[], yticks=[], title='receptive field' )
ax.plot( data.k_stim_true, lw=4, label='true', color='k', alpha=0.5 )
ax.plot( ml.k__d[:-1], color='g', lw=2, label='ml fit' )
ax.plot( ridge.k__d[:-1], color='r', lw=2, label='ridge fit' )
ax.legend( loc='lower right', fontsize=16 )
pass
```

```python
# ML overfits to the training data, makes poorer predictions on the testing data
print 'training: %.3f' % (ridge.LL_training_per_observation - ml.LL_training_per_observation)
print 'testing:  %.3f' % (ridge.LL_testing_per_observation - ml.LL_testing_per_observation)
```

# GLM with history dependence

## source data

```python
# instantiate
data = simulation.SimulatedData2DHistory( N_sec=60 )
D_history = data.D_history
```

## solving

```python
# maximum likelihood solution
ml = glm.ML( data, testing_proportion=0.2 )
ml.solve( verbose=False )

# ridge penalised solution
ridge = glm.Ridge( data, testing_proportion=0.2 )
ridge.training_slices = ml.training_slices
ridge.testing_slices = ml.testing_slices
ridge.solve( verbose=False )

# ALDs solver for history
alds = glm.Ridge_Stim_ALDs_History( data, testing_proportion=0.2 )
alds.training_slices = ml.training_slices
alds.testing_slices = ml.testing_slices
alds.solve( verbose=False )
```

```python
# plot history kernels
fig = plt.figure()
ax = fig.gca( xlabel='position', ylabel='value', xticks=[], yticks=[], title='history kernel' )
ax.plot([0, D_history], [0, 0], 'k-')
ax.plot( data.k_history_true, lw=4, label='true', color='k', alpha=0.5 )
ax.plot( ml.k[-D_history-1:-1], color='g', lw=2, label='ml fit' )
ax.plot( ridge.k[-D_history-1:-1], color='r', lw=2, label='ridge fit' )
ax.plot( alds.k__d[-D_history-1:-1], color='c', lw=2, label='ALDs fit' )
ax.legend( loc='lower left', fontsize=16 )
pass
```

```python
# plot receptive fields
fig = plt.figure()
ax = fig.add_subplot( 1, 4, 1, xlabel='position', ylabel='value', xticks=[], yticks=[], title='true' )
ax.imshow( data.k_true[:100].reshape(10, 10), cmap=plt.cm.Greys_r, vmin=-0.2, vmax=1 )
ax.set_aspect('equal')

ax = fig.add_subplot( 1, 4, 2, xlabel='position', ylabel='value', xticks=[], yticks=[], title='ML estimate' )
ax.imshow( ml.k[:100].reshape(10, 10), cmap=plt.cm.Greys_r, vmin=-0.2, vmax=1 )
ax.set_aspect('equal')

ax = fig.add_subplot( 1, 4, 3, xlabel='position', ylabel='value', xticks=[], yticks=[], title='ridge estimate' )
ax.imshow( ridge.k[:100].reshape(10, 10), cmap=plt.cm.Greys_r, vmin=-0.2, vmax=1 )
ax.set_aspect('equal')

pass
```

# Example from pre-existing data

## Data object

```python
# suppose we have this data (you can import your own)
source_data = simulation.SimulatedData2DHistory( N_sec=10, add_constant=False )
X_stim = source_data.X_stim
y = source_data.y__t

# note its shape
T, D_stim = X_stim.shape
print X_stim.shape
```

```python
# add a history to the regression matrix, extending 60 time steps back
D_history = 60
X_history = glm.construct_X_history( y, D_history )
```

```python
# construct the design matrix (excluding the column of ones)
X = glm.combine_X( X_stim, X_history, add_constant=False )

# note its shape
print X.shape
```

```python
# create the data structure
data = glm.Data( X, y )

# note that the column of ones has been added
print data.X.shape
```

## Solving

### ridge

```python
# now solve
ridge = glm.Ridge( data, solve=True, verbose=False )
```

```python
# parse the components of the kernel
k_stim = ridge.k[ :D_stim ]
k_history = ridge.k[ D_stim : D_stim + D_history ]
k_const = ridge.k[ -1 ]
assert len( ridge.k ) == D_stim + D_history + 1

# plot
fig = plt.figure( figsize=(12, 3) )
ax = fig.add_subplot( 1, 3, 1, xlabel='position', ylabel='value', xticks=[], yticks=[], title='fitted rf' )
ax.imshow( k_stim.reshape(10, 10), cmap=plt.cm.Greys_r, vmin=-0.2, vmax=1 )
ax.set_aspect('equal')

ax = fig.add_subplot( 1, 2, 2, xlabel='time (samples)', ylabel='value', xticks=[0, D_history], xticklabels=[-D_history, 0], title='history kernel', yticks=[] )
ax.plot( k_history, 'b', lw=2 )
ax.plot( [0, D_history], [0, 0], 'k-', alpha=0.5 )
pass
```

### ALDs

```python
# ALDs version
data.D_stim = D_stim
data.D_history = D_history
alds = glm.Ridge_Stim_ALDs_History( data, solve=True, verbose=False )
```

```python
# parse the components of the kernel
k_stim = alds.k[ :D_stim ]
k_history = alds.k[ D_stim : D_stim + D_history ]
k_const = alds.k[ -1 ]

# plot
fig = plt.figure( figsize=(12, 3) )
ax = fig.add_subplot( 1, 3, 1, xlabel='position', ylabel='value', xticks=[], yticks=[], title='fitted rf' )
ax.imshow( k_stim.reshape(10, 10), cmap=plt.cm.Greys_r, vmin=-0.2, vmax=1 )
ax.set_aspect('equal')

ax = fig.add_subplot( 1, 2, 2, xlabel='time (samples)', ylabel='value', xticks=[0, D_history], xticklabels=[-D_history, 0], title='history kernel', yticks=[] )
ax.plot( k_history, 'b', lw=2 )
ax.plot( [0, D_history], [0, 0], 'k-', alpha=0.5 )
pass
```

# Basis functions

## exponentials for history

```python
# ten basis functions as exponentials with different time constants
def basis( tau ):
    b = np.exp( np.arange( -D_history, 0 ) / float(tau) )
    return b / np.sum(b)

taus = np.logspace(0, np.log10(20), 10)
B = A([ basis(tau) for tau in taus ])
```

```python
# showing the bases
ax = gca()
for tau, b in zip( taus, B ):
    ax.plot( np.arange( -D_history, 0 ), b, label=r'$\tau = %.1f$' % tau )
ax.legend(loc='upper left')
pass
```

```python
# project into basis
X_history_in_basis = X_history.dot(B.T)
```

```python
# solve
X = glm.combine_X( X_stim, X_history_in_basis )
data = glm.Data( X, y )
ridge = glm.Ridge( data, solve=True, verbose=False )
```

```python
# parse the components of the kernel
k_stim = ridge.k[ :D_stim ]
k_history_in_basis = ridge.k[ D_stim : -1 ]
k_history = k_history_in_basis.dot(B)
k_const = ridge.k[-1]

# plot
fig = plt.figure( figsize=(12, 3) )
ax = fig.add_subplot( 1, 3, 1, xlabel='position', ylabel='value', xticks=[], yticks=[], title='fitted rf' )
ax.imshow( k_stim.reshape(10, 10), cmap=plt.cm.Greys_r, vmin=-0.2, vmax=1 )
ax.set_aspect('equal')

ax = fig.add_subplot( 1, 2, 2, xlabel='time (samples)', ylabel='value', xticks=[0, D_history], xticklabels=[-D_history, 0], title='history kernel', yticks=[] )
ax.plot( k_history, 'b', lw=2 )
ax.plot( [0, D_history], [0, 0], 'k-', alpha=0.5 )
pass
```
